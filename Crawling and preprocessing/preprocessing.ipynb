```python
import pandas as pd
import numpy as np
import os
import datetime
```

```python
def value_to_float(x):
    if str(x) == 'nan':
        return 0
    a = x.replace(',','')
    return np.float(a)
```

```python
data_file = os.listdir('Dataset')
data_file
```

```python
Start_day = datetime.datetime(2021,11,27)
for filename in data_file:
    Country = pd.read_csv(f'Dataset/{filename}', sep='\t')[:-7]
    for item in Country:
        try:
            if item == '#' or item == 'Continent' or item == None or item == 'Country':
                continue
            Country[item] = Country[item].apply(value_to_float)
        except:
            print(item, filename)
    Country['Date'] = Start_day
    Country.to_csv(f'Modified/Countries/{filename}', sep='\t', index=False)
    Start_day = Start_day + datetime.timedelta(days=1)
```

```python
Start_day = datetime.datetime(2021,11,27)
for filename in data_file:
    Continent = pd.read_csv(f'Dataset/{filename}', sep='\t')[-7:-1]
    keep_cols = ['Country', 'Total Cases', 'New Cases', 'Total Deaths', 'New Deaths', 'Total Recover', 
                            'New Recorver', 'Active Cases', 'Serious Critical']
    Continent = Continent[keep_cols]
    Continent = Continent.rename(columns={"Country": "Continent"})
    for item in Continent:
        try:
            if item == '#' or item == 'Continent' or item == None or item == 'Country':
                continue
            Continent[item] = Continent[item].apply(value_to_float)
        except:
            print(item, filename)
    Continent['Date'] = Start_day
    Continent.to_csv(f'Modified/Continents/{filename}', sep='\t', index=False)
    Start_day = Start_day + datetime.timedelta(days=1)
```

```python
Start_day = datetime.datetime(2021,11,27)
for filename in data_file:
    World = pd.read_csv(f'Dataset/{filename}', sep='\t').tail(1)
    keep_cols = ['Total Cases', 'New Cases', 'Total Deaths', 'New Deaths', 'Total Recover', 
                            'New Recorver', 'Active Cases', 'Serious Critical','Cases/1M pop','Deaths/1M pop']
    World = World[keep_cols]
    for item in World:
        try:
            World[item] = World[item].apply(value_to_float)
        except:
            print(item, filename)
    World['Date'] = Start_day
    Start_day = Start_day + datetime.timedelta(days=1)
    World.to_csv(f'Modified/Worlds/{filename}', sep='\t', index=False)
```
